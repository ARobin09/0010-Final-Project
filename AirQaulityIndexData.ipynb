# Air Quality Index Introduction
One metric we chose was air quality. Air quality is pretty important as cleaner air has better health benefits in comparison to smog. It is also a very obvious indicator of cleanliness.

```python
import pandas as pd
```

```python
df = pd.read_csv('https://data.wprdc.org/datastore/dump/4ab1e23f-3262-4bd3-adbf-f72f0119108b')

ave_aqi = df.groupby('site')['index_value'].mean()

ave = ave_aqi.drop('Liberty 2').drop('Lawrenceville 2').reset_index()

ave_aqi_filtered = ave.sort_values(by='index_value', ascending = True).reset_index()

ave_aqi_final = ave_aqi_filtered.drop(columns=['index'])

ave_aqi_final.columns = ['neighborhood', 'index_value']

print(ave_aqi_final)
```

```python
import matplotlib.pyplot as plt

ave_aqi_final_sorted = ave_aqi_final.sort_values(by='index_value', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(ave_aqi_final_sorted['neighborhood'], ave_aqi_final_sorted['index_value'], color='seagreen')
plt.xlabel('Average AQI', fontsize=13)
plt.ylabel('Location', fontsize=13)    
plt.title('Average Air Quality Index by Location (Lower is better)', fontsize=20)
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
plt.show()
```

# Analysis/Conclusion
Based on the above data, we can see that the averages of the above neighborhoods are below 50 in the AQI index, which is the safe zone for air quality. The cleanest neighborhoods in terms of air quality are West Mifflin, FLag Plaza, Glassport High Street, Liberty, and North Braddock. These 5 neighborhoods all have had the lowest average AQI in their data, indicating that they were the cleanest cities compared to the other 13 in the graph above. Note that Lawrenceville 2 and Liberty 2 were dropped from the data set, due to being different measurements of PPI, giving an unfair advantage to both cities.
