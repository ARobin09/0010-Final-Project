Green spaces are areas of vegetation in an urban environment. THey are filled with plants and are accessible to the public. Green spaces are a good indicator of cleanliness because they reduce the neighborhood's carbon footprint.


```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("https://data.wprdc.org/dataset/cd9ae013-5b93-43d6-8793-bccf10f12c72/resource/b06fa202-ead7-491c-b8f6-2702dd35f8dd/download/opendata-pubworks-operations-green-spaces-2015.csv")

spaces = data['Neighborhood'].value_counts().reset_index()

spaces.columns = ['neighborhood','number of spaces']

spaces['neighborhood'] = spaces['neighborhood'].replace('East Liberty', 'Liberty')
 
print(spaces)
```

```python
graph = spaces.plot(kind='barh', x='neighborhood', y='number of spaces', legend=False, width=0.6, figsize = (10,5), color='seagreen')

plt.xlabel('Neighborhood', fontsize=15)
plt.ylabel('Number of Green Spaces', fontsize=15)
plt.title('Green Spaces by Neighborhood', fontsize=30)

graph.tick_params(axis='x', labelsize=12)
graph.tick_params(axis='y', labelsize=12)
graph.invert_yaxis()

plt.show()
```

# Analysis/Conclusion
The data above indicates that the neighborhoods with the most green spaces are Liberty, Point Breeze, and Central Business District. Note 'East Liberty' was changed to 'Liberty' to make merging all the data in the final notebook easier.
