Smart trash cans have sensor that detect fill-level data, which is then sent to waste management services. Smart trash cans are a good indicator of cleanliness because it shows these neighborhoods are environmentally-concious.

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

data = pd.read_csv("https://data.wprdc.org/datastore/dump/75b83ac9-8069-4cf1-bcc3-b9e6b04487d9")

locations = data['neighborhood'].value_counts().reset_index()

locations.columns = ['neighborhood','number of cans']

locations['neighborhood'] = locations['neighborhood'].replace('East Liberty', 'Liberty')

print(locations)
```

```python
graph = locations.plot(kind='barh', x='neighborhood', y='number of cans', legend=False, width=0.6, figsize = (20,25), color='seagreen')

plt.xlabel('Neighborhood', fontsize=30)
plt.ylabel('Number of Cans', fontsize=30)
plt.title('Smart Cans by Neighborhood', fontsize=50)

graph.tick_params(axis='x', labelsize=15)
graph.tick_params(axis='y', labelsize=15)
graph.invert_yaxis()

plt.show()
```

# Analysis/Conclusion
The data above indicates that the neighborhoods with the most smart trash cans are Shadyside, Liberty, Squirrel Hill South, Bloomfield, and North Shore. Shadyside has the most with 76, while North Shore has 53. Note 'East Liberty' was changed to 'Liberty' to make merging all the data in the final notebook easier.
